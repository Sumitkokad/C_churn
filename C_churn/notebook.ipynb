Machine learning Phase

```python
# Feature engineering is starts -->
```

```python
df.columns
```

```python
y=df[['Churn']]
```

```python
y
```

```python
x=df[['Age','Gender','Tenure','MonthlyCharges']]
```

```python
x
```

```python
# Encode the columns in the x and y data column like Gender and churn convert into numeric value
```

```python
# For y churn means = 1 and not churn means = 0
y['Churn']=y['Churn'].apply(lambda x : 1 if x=='Yes' else 0)
```

```python
y
```

```python
# Now encode the x columns like Gender for female = 1 for Male = 0

x['Gender']=x['Gender'].apply(lambda x: 1 if x=='Female' else 0)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
# We use scale the data using standard scaler
from sklearn.preprocessing import StandardScaler
```

```python
# Initialize the standard scaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(x_train)
```

```python
# # Save the trained scaler object to a file so it can be reused later without retraining
import joblib
joblib.dump(scaler,'scaler.pkl')
```

```python
# X test values are also scaled

X_test=scaler.fit_transform(x_test)
```

```python
X_train
```

```python
X_test
```

*Now Both the training and testing data is scaled and we can use this data to train our model and make predictions*

```python
# This function is used to check the performance of the model
# Importing the classification model
from sklearn.metrics import accuracy_score
# It shows the accuracy in 0 and 1-> 1 is better than 0
# This function is used to check the performance of the model by calculating the accuracy score and printing it in a formatted way
def modelperformance(predictions):
    print("Accaracy Score of the model is: {:.2f}".format(accuracy_score(y_test, predictions)))
```

```python
# We already check the best model is support vector machine for this dataset and this working so we import basially the support vector machine algorithm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
import joblib
```

```python
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
```

```python
svm = SVC(probability=True, class_weight='balanced')

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svm = GridSearchCV(svm, param_grid, cv=5, scoring='f1')

grid_svm.fit(X_train_res, y_train_res)

best_model = grid_svm.best_estimator_
```

```python
y_pred = best_model.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
modelperformance(y_pred)
```

```python
# model saved sucessfully
joblib.dump(best_model, 'best_model.pkl')

print("\nModel saved successfully ✅")
```
